## 教師なし学習とは?

機械学習の手法の一つです。  
教師なし学習は元のデータに正解が無く、それを導き出す手法のことをいいます。  
「出力すべきもの」があらかじめ決まっていないという点で教師あり学習とは大きく異なります。  
教師なし学習は膨大なデータを自動的に解析させますが、正解がないので算出した特徴量から構造、法則、傾向、分類、定義などを導き出します。  
メリットとして傾向分析、未来予測などにも応用できる事や機械自身が特徴や定義を発見するため、データが膨大にあれば、ラベル付きデータを用意する必要や手間がない点です。  
デメリットとして、k-meansなどクラスタリングでよく使われる手法では自ら分けるクラスタの数（分類する数)を決めなければならない点です。  

教師なし学習には,  
**クラスタリング**  
・階層型クラスタリング(ユークリッド距離＊ウォード法など)  
・非階層型クラスタリング（k-meansなど)  
・トピックモデル（LDAなど)  
**ニューラルネットワーク**  
・自己組織化マップ（SOM)  
などがあります。

## Feature Extraction
次に**PCA(主成分分析)**を行いましょう。  
PCAは主に次元削減を行う際に行います。  
「次元削減」とは、文字通り、データの次元数を減らすことです。  
ここでいう「次元数」は、データセットのフィーチャーの数と言い換えることができます。  
次元削減を行う目的は主に以下の二つです。  
・データの圧縮  
・データの可視化  
データセットが膨大になりやすい機械学習の分野において、このデータの圧縮は計算資源の有効活用という点について、非常に有用です。  
xとyの二つについて計算を行っていたところをz一つに対する計算で済ませることができるわけですから、単純に、高速に計算を行うことができるようになります。  

またデータの可視化ですが、4次元以上の情報を端的に表現することは難しいと言わざるを得ません。  
そこで行うのが「次元削減」です。  
上のプロットのように、２次元の可視化で表現するのが普通ですが次元圧縮することで1度に４次元以上の特徴を可視化することが可能になります。  

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(train_fs)
```

```python
pca_train = pca.transform(train_fs)
```

```python
pca_train[:10]
```

```python
plt.scatter(pca_train[:,0],pca_train[:,1])
```

次元圧縮したデータに意味を探すのは難しいですが、上のプロットは全てのデータがうまく次元削減できて、 中央によっているということがわかりました。  
実際に累積寄与率(分散説明率)を見て確認して見ましょう。

```python
print('累積寄与率(分散説明率): {0}'.format(sum(pca.explained_variance_ratio_)))
```

累積寄与率は約 0.71 で、ようするに第二主成分までで元のデータの 71% が表現できていることが分かります。

# 3.クラスタリング  
最初に説明したようにクラスタリングは教師なし学習の手法の一つで主に、答えの無いデータをクラス分けする時に使います。  
ただクラス分けと言いましたが、クラス分類と違い、元々決まったクラスがありそれを分けるわけではありません。  
イメージですが、身長と体重といったデータがあった場合決まったクラスはないですが、クラスタリングをすることで近い特徴を持った人を分けて考えることができます。  
身長と体重から2つにクラス分けした場合、デブかガリガリかに分けることができると言えばイメージできるのではないでしょうか？

クラスタリングには階層的手法と非階層的手法があります。  
非階層的手法は、代表的な物でk-means（k-平均法）があります。  
クラスタ(分けるクラス)の平均の値を考えながら、指定したk個のクラスタに分類するためにこのような名前がついています。  
ランダムに、クラスタを置きそのクラスタから最も近い距離を計算し分けられる中心点を探すという手法です。  
非階層的手法は自分の手でクラスタの数を決めないといけないというデメリットがあります。  

階層的クラスタリングは、名前の通り階層に分けていったり、もしくは近い特徴同士をくっつけていきながら適切なクラスタを見つける手法です。  
階層的クラスタリングには凝集型と分割型があり、代表的なものとして凝集型と呼ばれるもののひとつに完全連結法があります。  
k-means法と完全連結法はそれとは全く違うアプローチです。  
k-means法のように、クラスタ数kは関係ありません。  
まずは、全てのサンプルは、別々のクラスタに所属していると考えるのです。そして、距離の近いクラスタを順番に仲間にしていく手法です。

まずは階層的手法を使いクラス分けし可視化していきます。

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
linkage_result = linkage(train_fs, method='ward', metric='euclidean')
# 閾値を決める
threshold = 0.5 * np.max(linkage_result[:, 2])
```

```python
plt.figure(num=None, figsize=(16, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(linkage_result, color_threshold=threshold)
plt.show()
```

```python
clustered = fcluster(linkage_result, threshold, criterion='distance')
```

```python
print(clustered)
```

3種類に分けることができました。  
このように階層的クラスタリングの凝集型は、近いもの同士をつなぎ合わせ最後は一つのクラスになるような手法で、どのように分けているか可視化することが可能です。

次は非階層的クラスタリングのK-meansを試していきましょう。  
K-meansはまず分けるクラスタ数を探す必要があります。  
その方法としてエルボー法やシルエット分析があります。

```python
from sklearn.cluster import KMeans
SSE = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, max_iter=300, tol=1e-04, random_state=1)
    km.fit(train_fs)
    SSE.append(km.inertia_)

plt.plot(range(1,11), SSE, marker='o')
plt.xlabel('k')
plt.ylabel('SSE')
plt.show()
```

```python
km = KMeans(n_clusters=2, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(train_fs)
from matplotlib import cm
from sklearn.metrics import silhouette_samples
import numpy as np
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(train_fs, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(i / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2)
    y_ax_lower += len(c_silhouette_vals)

silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--")
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.show()
```

```python
km = KMeans(n_clusters=3, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(train_fs)
from matplotlib import cm
from sklearn.metrics import silhouette_samples
import numpy as np
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(train_fs, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(i / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2)
    y_ax_lower += len(c_silhouette_vals)

silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--")
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.show()
```

エルボー図では横軸がクラス数になります。  
一番変化が少なくなるところをクラス数とします。  
上の図では２か３種類に分けるのが良さそうだとわかります。  

シルエット図は、各クラスタのサンプルの凝集度（まとまり具合）を表示して、それぞれを比べます。  
上で見ると、2種分類も種分類もうまくできています。  
しかし3種分類の黄色いグラフ(1つのクラスタ)だけ少しうまく分類できていないことも確認できました。

# オプション・k-meansスクラッチ

```python
from numpy.random import randint
```

```python
def random_point(x,k):
    np.random.seed(5)
    r1= randint(min(x[:,0]),max(x[:,0]),(1,k))
    r2 = randint(min(x[:,1]),max(x[:,1]),(1,k))
    point = np.r_[r1,r2]
    return point
```

```python
def k_mean(x,k,ite):
    x = np.matrix(x)
    point = random_point(x,k).T
    crassta = randint(0,k, x.shape[0])
    norm = 0
    norm2 = np.array([[0.0,0.0]])
    count = 0
    #norm_list = [[[],[],[]]]
    norm_list = [[] for i in range(k)]
    for i in range(ite):
        norm_list = [[] for i in range(k)]
        for i in range(k):
            for e,d in enumerate(x):
                norm += np.linalg.norm(x[e]-point[i])
                norm_list[i].append(np.linalg.norm(x[e]-point[i]))
        classta_list =  np.array(norm_list).argmin(0)
        for i in range(k):
            for a,b in zip(x,classta_list):
                if b == i:
                    norm2 +=  a
                    count += 1
            norm2 /= count
            point[i] = norm2
            norm2 = np.array([[0.0,0.0]])
            count = 0
               
    return np.array(norm_list).argmin(0),point.T
```
